```python
# -*- coding: utf-8 -*-
"""
@Author: Guanfu Liu
To plot the figure of z versus t
"""

import matplotlib.pyplot as plt
from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
import numpy as np
from astropy.cosmology import z_at_value
%matplotlib notebook


cosmo = FlatLambdaCDM(H0=70*u.km/u.s/u.Mpc, Om0=0.3)
ages = np.linspace(0.2,13,100)*u.Gyr
z = np.array([z_at_value(cosmo.age, age) for age in ages])
fig, ax = plt.subplots(nrows=1,ncols=1,dpi=100)
ax.plot(z,13-ages.value)
ax.set_xlabel('z(redshift)')
ax.set_ylabel('t(Gyr)')
fig.show()
```
