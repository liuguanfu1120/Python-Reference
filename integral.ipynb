```python
import mpmath as mpm

def func(x):
    a0,a1,a2,a3,a4 = 1,1,1,1,10
    return mpm.exp(-0.5*(a4*x**4+a3*x**3+a2*x**2+a1*x+a0))

```

```python
mpm.dps=50
mpm.pretty=True
print(mpm.quad(func,[-1,1], maxdegree=10))
print(mpm.quad(func,[-10,10], maxdegree=10))
print(mpm.quad(func,[-100,100], maxdegree=10))
print(mpm.quad(func,[-1000,1000], maxdegree=100))
print(mpm.quad(func,[-mpm.inf,mpm.inf], maxdegree=10))
```
