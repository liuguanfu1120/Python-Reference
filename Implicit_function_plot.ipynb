```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
delta = 0.001
rho = np.arange(0.0001, 1.0, delta)
z = np.arange(0.0001, 1.0, delta)
X, Y = np.meshgrid(rho, z)
A, B = 1, -1
f = A + B*Y/np.sqrt(X**2+Y**2)
f_all = [0.05,0.1,0.2,0.5]
fig, ax = plt.subplots(nrows=1,ncols=1,dpi=100)
CS = ax.contour(X, Y, f, levels = f_all,colors = ['black','black','black','black'])
plt.clabel(CS,fontsize=10)
# plt.show()
```
