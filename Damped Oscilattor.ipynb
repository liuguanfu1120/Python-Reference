```python
# -*- encoding: utf-8 -*-
# @Author: Liu Guanfu


from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

def pend(y, t):
    y1, y2 = y
    dydt = [y2, -y1-y2]
    return dydt

y0 = np.array([1,-1/2])
t = np.linspace(0,10,100)
sol = odeint(pend, y0, t)

x = sol[:,0]
velocity = sol[:,1]

x1 = np.exp(-t/2)*np.cos(np.sqrt(3)/2*t)

velocity1 = -np.exp(-t/2)*np.cos(np.sqrt(3)*t/2)/2 - np.sqrt(3)*np.exp(-t/2)*np.sin(np.sqrt(3)*t/2)/2


fig,ax = plt.subplots(nrows=1,ncols=1,dpi=120)
ax.plot(t,x,label=r'Numerical results')
ax.plot(t,x1,'-.',label=r'Analytical results')
ax.legend()
ax.set_title('Displacement')
ax.set_xlabel(r'Time/s')
ax.set_ylabel(r'x/m')
ax.grid()
fig.show()

fig,ax = plt.subplots(nrows=1,ncols=1,dpi=120)
ax.plot(t,velocity,label=r'Numerical results')
ax.plot(t,velocity1,'-.',label=r'Analytical results')
ax.legend()
ax.set_title('Velocity')
ax.set_xlabel(r'Time/s')
ax.set_ylabel(r'v/(m/s)')
ax.grid()
fig.show()
```
